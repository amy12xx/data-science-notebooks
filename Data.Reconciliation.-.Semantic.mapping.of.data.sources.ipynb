# Data reconciliation: Semantic mapping of disparate data sources

This is part of the data reconciliation solution. The first step towards data recon is to predict matching columns from different data sources, by analyzing the data set, and the match rate for the suggested mappings.

## Approach

Learning Source Descriptions (LSD) is used to perform semantic mapping of disparate data sources. LSD, briefly, is used to first train different learners, combine the learnings into a meta-learner to apply on the final test data. 
For more details, please check the source paper: http://homes.cs.washington.edu/~pedrod/papers/sigmod01.pdf

### Training Learners used:
Naive Bayes learner: trains on textual data using Naive Bayes approach
Datatype Leaner: trains on data type of data using Random Forest algorithm

### Meta Learner used:
Logistic Regression is used for the meta learner to learn the weights of the individual learners

Naive Bayes approach is useful for textual data, while datatype learner is useful when datatype is strongly suggestive of data (such as datetime datatype etc.). 
Other learners can also be implemented to enhance the model. 
Further domain constraints and user feedback can also be incorporated into the model.
