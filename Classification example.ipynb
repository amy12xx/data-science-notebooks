# Example of Classification problem on Iris dataset

```python
# import required libraries
% matplotlib inline

from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load example (classification) dataset
iris = datasets.load_iris()
```

```python
# convert to pandas dataframe to get some more information about data
irisdf = pd.DataFrame(iris.data)
```

```python
# Example plots
# Plots histogram of first column
sns.distplot(irisdf[0])
```

```python
# View relationship between features
sns.pairplot(irisdf)
```

```python
# import classifier
from sklearn import svm
```

```python
# define classifier 
clf = svm.SVC()
X, y = iris.data, iris.target
```

```python
# import libraries to split data set
from sklearn.cross_validation import train_test_split
```

```python
# split dataset into 80:20 random train and test subsets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# train classifier on train set
clf.fit(X_train, y_train)
```

```python
# predict on test set
predictions = clf.predict(X_test)
```

```python
# Import libraries to test performance of model
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# print score
accuracy_score(y_test, predictions)
```

```python
# check confusion matrix
confusion_matrix(y_test, predictions)
```
